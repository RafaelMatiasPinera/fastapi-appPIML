```python
# Exploración básica del DataFrame Output
import pandas as pd

df_output = pd.read_csv('output.csv')
df_output.head()
df_output.info()
df_output.describe()
```

```python
#VOY A CORROBORAR QUE ITEM_ID DE AUSTRALIAN ITEMS SE CORRESPONDA CON ID DE OUTPUT
unique_items_australian = df_items['item_id'].unique()
unique_items_output = df_output['item_id'].unique()

# Verificar si todos los elementos de items_df['item_id'] están en output_df['item_id']
all_items_in_output = all(item in unique_items_output for item in unique_items_australian)

if all_items_in_output:
    print("Todos los 'item_id' en items.csv están en output.csv.")
else:
    print("Algunos 'item_id' en items.csv no están en output.csv.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Histograma de una variable numérica
# Histograma de una variable numérica
plt.figure(figsize=(10, 6))
sns.histplot(df_output['price'], bins=30, kde=True)
plt.title('Distribución de Precios')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Crear una nueva columna 'genre' que contenga los géneros individuales
df_output['genre'] = df_output['genres'].apply(lambda x: [genre.strip() for genre in x.strip('[]').replace("'", "").split(",")] if pd.notnull(x) else [])

# Convertir la lista de géneros en filas separadas
df_genres = df_output['genre'].explode().reset_index(drop=True)

# Visualización de la frecuencia de géneros
plt.figure(figsize=(14, 8))
sns.countplot(y=df_genres, order=df_genres.value_counts().index)
plt.title('Frecuencia de Géneros')
plt.xlabel('Frecuencia')
plt.ylabel('Género')
plt.show()
```

```python
# Calcular el promedio de los precios de los juegos
precio_promedio = df_output['price'].mean().round()

# Imprimir el resultado
print("El precio promedio de los juegos es USD:", precio_promedio)
```

```python
import matplotlib.pyplot as plt

# Crear un histograma de la variable 'playtime_forever'
plt.figure(figsize=(10, 6))
plt.hist(df_items['playtime_forever'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribución de Tiempo de Juego')
plt.xlabel('Tiempo de Juego')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calcular la frecuencia de cada valor en la columna Sentiment_analysis
sentiment_counts = df_reviews['Sentiment_analysis'].value_counts()

# Crear un gráfico de barras
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribución del Análisis de Sentimiento')
plt.xlabel('Sentimiento')
plt.ylabel('Frecuencia')
plt.xticks(rotation=0)  # Rotar etiquetas en el eje x si es necesario
plt.grid(axis='y')
plt.show()
```
